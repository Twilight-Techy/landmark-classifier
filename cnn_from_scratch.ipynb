- How does your code resize the images (by cropping, stretching, etc)?  What size did you pick for the input tensor, and why?

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> **Answer**: 

I preprocess images by first resizing the shorter side to 256 and then cropping to a fixed square input of 224×224. For training I use RandomCrop(224) (after Resize(256)) together with augmentation: RandomHorizontalFlip, small RandomRotation (±10°) and modest ColorJitter (brightness/contrast/saturation/hue). Validation and test use Resize(256) + CenterCrop(224) with no augmentation.
I normalize images using the dataset mean and standard deviation computed by compute_mean_and_std() (this produces per-channel mean/std and is cached), and apply transforms.ToTensor() before normalization.
Rationale: 224×224 is a standard input size (compatible with common CNN architectures) and fits well in GPU memory. Resizing to 256 then cropping preserves aspect ratio and avoids stretching; random cropping, flips, small rotations and color jitter increase robustness to viewpoint/lighting variations and reduce overfitting on a relatively small dataset. I did not apply extreme geometric transforms (large rotations or heavy warps) because they would distort landmarks. If class imbalance is observed, reasonable next steps would be oversampling low-frequency classes, using class-weighted loss, or stronger augmentation targeted at small classes.

<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question**: Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ 

I started with a simple, well-known CNN pattern: repeated small (3×3) convolutional blocks followed by pooling. Each block uses Conv → ReLU → BatchNorm to give nonlinearity and stable, faster training. I progressively increase channel depth (32 → 64 → 128 → 256 → 512) as spatial dimensions shrink — this preserves capacity while reducing spatial resolution, which is efficient for feature learning. I use MaxPool2d for spatial downsampling and AdaptiveAvgPool2d((1,1)) to produce a fixed-length descriptor regardless of input spatial size (useful for 224×224 inputs). The classifier is a compact MLP with two linear layers and Dropout for regularization; the final linear layer uses the num_classes parameter so the model is flexible for different tasks.

**Design choices rationale:**
(1) small 3×3 kernels stack well and are parameter-efficient; 
(2) BatchNorm and Dropout help generalization and training stability; 
(3) Adaptive pooling avoids hardcoding a flattened spatial size; 
(4) the network is intentionally moderate in depth so it can be trained from scratch within reasonable time on the dataset and reach a sensible baseline before trying transfer learning.

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

````python
batch_size = 32        # bigger batches -> smoother gradients
valid_size = 0.2       # keep this the same
num_epochs = 50        # still 50, you can early stop if needed
num_classes = 50       # do not change this
dropout = 0.3          # reduce dropout (less underfitting)
learning_rate = 0.001  # good default for Adam
opt = 'adam'           # switch to Adam optimizer
weight_decay = 1e-4    # small regularization to prevent overfitting
````

````python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel
from src.workspace_utils import active_session
import os

# get the data loaders using batch_size and valid_size defined in the previous
# cell (use all CPUs for num_workers)
data_loaders = get_data_loaders(batch_size=batch_size, valid_size=valid_size, num_workers=-1)

# instance model MyModel with num_classes and dropout defined above
model = MyModel(num_classes=num_classes, dropout=dropout)

# Get the optimizer using get_optimizer and the model you just created
optimizer = get_optimizer(model, optimizer=opt, learning_rate=learning_rate, weight_decay=weight_decay)

# Get the loss using get_loss
loss = get_loss()

# ensure checkpoint directory exists so saving won't fail
os.makedirs("checkpoints", exist_ok=True)

with active_session():
    optimize(
        data_loaders,
        model,
        optimizer,
        loss,
        n_epochs=num_epochs,
        save_path="checkpoints/best_val_loss.pt",
        interactive_tracking=True
    )
````

Allright, now we are ready to export our model using our Predictor class:

````python
# NOTE: you might need to restart the notebook before running this step
# If you get an error about RuntimeError: Can't redefine method: forward on class
# restart your notebook then execute only this cell
from src.predictor import Predictor
from src.helpers import compute_mean_and_std
from src.model import MyModel
from src.data import get_data_loaders
import torch

data_loaders = get_data_loaders(batch_size=1)

# First let's get the class names from our data loaders
class_names = data_loaders["train"].dataset.classes

# Then let's move the model_transfer to the CPU
# (we don't need GPU for inference)
model = MyModel(num_classes=50, dropout=0.5).cpu()

# Let's make sure we use the right weights by loading the
# best weights we have found during training
# NOTE: remember to use map_location='cpu' so the weights
# are loaded on the CPU (and not the GPU)

ckpt_path = "checkpoints/best_val_loss.pt"
ckpt = torch.load(ckpt_path, map_location="cpu")

if isinstance(ckpt, dict):
    # common wrapper keys
    if "model_state" in ckpt:
        state_dict = ckpt["model_state"]
    elif "state_dict" in ckpt:
        state_dict = ckpt["state_dict"]
    else:
        # assume it's already a state_dict
        state_dict = ckpt
else:
    # if someone saved a plain state_dict
    state_dict = ckpt

model.load_state_dict(state_dict)
print(f"Loaded checkpoint from {ckpt_path}")

# Let's wrap our model using the predictor class
mean, std = compute_mean_and_std()
predictor = Predictor(model, class_names, mean, std).cpu()

# Export using torch.jit.script
scripted_predictor = torch.jit.script(predictor)

scripted_predictor.save("checkpoints/original_exported.pt")
````
